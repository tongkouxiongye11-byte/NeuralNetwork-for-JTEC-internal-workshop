# ・変数の使い方、標準出力(print)
### 変数への代入方法と、print関数を使った様々な出力方法の例

```python
a = 100         # aに100を代入
b = "Hello"     # bにHelloを代入
c = "World"     # cにWorldを代入

print(a)        # 「100」を出力
print(b,c)      # 「Hello World」を出力
print(b+c)      # 「HelloWorld」を出力
print("a =",a)  # 「a = 100」を出力
print(f"a={a}, {b} my {c}")  # 「a=100, Hello my World」を出力
```

# ・基本演算
### 数値計算や代入演算子の使い方

```python
a = 5
b = 10

# 基本的な四則演算
print("a+b =",a+b)  # 加算(足し算)
print("a-b =",a-b)  # 減算(引き算)
print("a*b =",a*b)  # 乗算(掛け算)
print("a/b =",a/b)  # 除算(割り算)
print("b%a =",b%a)  # 剰余(割り切った時の余り)

# 代入演算子
c = 10
c = c + a           # c(10)とa(5)を足した数を、新しくcに代入(=15)
print("c =",c)
c += 2              # c = c + 2 と同じ意味（現在の値に2を加算して更新）
print("c =",c)
```

# ・リスト
#### データの集合を扱う「リスト」(他の言語で言うところの配列)

```python
#--------
# リスト
#--------
a = [1,2,3,4,5]
print("a =", a)            # a(リスト)を表示
print("a[1] =", a[1])      # 1番目の要素を表示(インデックス指定)
print("a[1:3] =", a[1:3])  # 1番目から3番目の手前までを表示(スライス)

a.append(6)      # リストの末尾に要素を追加
print("a =", a)

a[1] = 100       # リストの内容を変更(1番目の要素を100に変更)
print("a =", a)

# --- リストには異なる型同士でも入れることができる ---
b = [[1,2], 3, "Hello"]  # リスト、数字、文字列を一つのリストに
print("b =",b)
```

# ・if、for
#### インデントが重要

```python
#--- if ：条件分岐 ---
#--- for：繰り返し ---

for n in range(10): # nが0〜9までの10回分繰り返す(range(10) は0から9までの連続した数値を生成)
  if n % 3 == 0:    # nを3で割った時の余りが0の時
    print(f"n={n}は3の倍数")
  if n >= 7:        # nが7以上の時
    print(f"n={n}は7以上です")
  if n == 5:        # nが5の時
    print(f"n={n}です")
```

# ・関数
#### 特定の処理をひとまとめにしたもの

```python
#--- 引数なし関数 ---
#--- Hello Youtubeを表示する ---
def Func_A():
    print("Hello Youtube")

#--- 引数あり関数 ---
#--- 二つの入力を足し合わせる ---
def Func_B(x,y):
    return x+y

Func_A()  # Func_Aを実施(コール)

a, b = 10, 20
print("Func_B(a,b) =", Func_B(a,b))   # Func_Bの実施結果を表示


#--- 関数内の演算が正しければ引数の型は問わない ---
n = [1,2,3]
m = [4,5,6]
print("Func_B(n,m) =",Func_B(n,m))  # リストの結合

s = "Python"
t = "学習"
print("Func_B(s,t) =",Func_B(s,t))  # 文字列の結合
```

# ・***Numpy***
#### 数値計算を行う上で必須のライブラリ

```python
import numpy as np   # numpyを「np」という名前でインポート（慣習的な書き方）

a = np.array([1,2,3])        # 1次元配列（ベクトル）
print("a =",a)               # a(ndarray)を表示
print("a[1] =",a[1])         # 1番目の要素を表示(リストとアクセスの仕方は同じ)
print("a.shape =",a.shape)   # 配列の形を表示
print("a.ndim =",a.ndim)     # 次元数を表示
print("a.size =",a.size)     # 全要素数を表示

b = np.array([[1,2,3],[4,5,6]])  # 2次元配列（行列）
print("b =",b)                   # b(ndarray)を表示
print("b[1][2] =",b[1][2])       # 1行目・2列目の要素にアクセス
print("b.shape =",b.shape)       # 配列の形を表示
print("b.ndim =",b.ndim)         # 次元数を表示
print("b.size =",b.size)         # 全要素数を表示

c = np.zeros((2,3))   # すべて0で埋められた2行3列の配列
print("c =",c)
d = np.ones(3)        # すべて1で埋められた要素数3の配列
print("d =",d)

e = np.random.random()     #0以上1未満の乱数を生成
print("e =",e)
f = np.random.randn(3,4)   # 標準正規分布に従う乱数で3行4列の配列を生成
print("f =",f)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
# Numpy配列同士は要素ごとの演算結果になる(サイズはそのまま)
print("a+b =",a+b)

# リストでは+演算は結合になる
a_list = [1,2,3]
b_list = [4,5,6]
print("a_list+b_list =",a_list+b_list)

c = np.array([[1,2],[3,4]])
# 1 2
# 3 4
d = np.array([[5,6],[7,8]])
print("c*d =",c*d)    # 乗算(掛け算)も同様

e = 10
print("a*e =",a*e)  # ブロードキャスト(全ての要素を10倍する）
                    # スカラーの10が配列のサイズに合わせて適当に拡張

#--- 入力を2倍にして返す関数 ---
def func_np(x):
  return x*2
print("func_np(c) =",func_np(c))   # 入力したNumpy配列に合わせて全ての要素が2倍されるようにブロードキャストされる

print("c.T =",c.T)   # 行と列を入れ替える(Transposition(転置)のT)

print("np.sum(c) =",np.sum(c))                # 配列の全要素を足した結果
print("np.sum(c,axis=0) =",np.sum(c,axis=0))  # 配列を行方向に足した結果
print("np.sum(c,axis=1)",np.sum(c,axis=1))    # 配列を列方向に足した結果
```

# ・matplotlib
#### グラフを描画するライブラリ

```python
# import numpy as np   #前のセルでインポートしてるのでなくても良い
import matplotlib.pyplot as plt   # 「plt」という名前で描画用のライブラリをインポート（慣習的な書き方）

x = np.linspace(-5,5)  # -5から5の間を等間隔に区切った50個の数値を作成
y = 2*x + 1            # 一次関数の計算

# --- 線グラフの描画 ---
plt.plot(x,y)     # xとyの値を線でつないで描画する
plt.grid()        # グラフにグリッド（方眼紙のような網目）を入れる
plt.show()        # 作成したグラフを表示する

# --- 散布図の描画 ---
plt.scatter(x,y)  # xとyの値を「点」としてプロットする
plt.show()        # 作成したグラフを表示する
```

# ・***クラス***  
#### データと機能をひとまとめにしたもの

```python
class Human:
    # --- コンストラクタ（初期化メソッド） ---
    # インスタンス（実体）が作られた時に自動的に実行されます
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # --- メソッド(クラス内で定義する関数) ---
    # コンストラクタ含め第一引数はselfを設定する
    # selfはインスタンス自身を指す特別な引数
    def Greeting(self):
        print(f"私の名前は{self.name}です。私の年齢は{self.age}歳です。")

    def HappyBirthday(self):
        self.age += 1     # selfでインスタンス変数にアクセス
        self.Greeting()   # selfでメソッドにアクセス

    def Attack(self, x):
        y = x*2
        print(f"威力{y}のパンチをくりだした。")


MURABITO_A = Human("村人A", 42)  # 村人Aというインスタンスの作成
MURABITO_A.Greeting()            # Greetingメソッドの実行
MURABITO_A.HappyBirthday()       # HappyBirthdayメソッドの実行
MURABITO_A.Attack(10)            # Attackメソッドの実行

MURABITO_B = Human("村人B", 36)  # 村人Bというインスタンスの作成
MURABITO_B.Greeting()
MURABITO_B.HappyBirthday()
MURABITO_B.Attack(20)
```

```python
#--------------
# クラスの継承
#--------------
class Hero(Human):   #HumanのクラスをベースにHeroのクラスを作成
    def Attack(self, x):   #Humanとは同じメソッドだが、別の処理を行う
        y = x*10
        print(f"威力{y}のパンチをくりだした。")

    def Ability(self):     #Hero固有のメソッド
        print("瞬間移動ができます")

YUSHA = Hero("勇者たかし", 50)  # 勇者たかしというインスタンスの作成
YUSHA.Greeting()                # Greetingメソッドの実行
YUSHA.HappyBirthday()           # HappyBirthdayメソッドの実行
YUSHA.Attack(10)                # Attackメソッドの実行
YUSHA.Ability()                 # Abilityメソッドの実行
```
