```python
#------
# Q2-1
#------
import numpy as np
import matplotlib.pyplot as plt

def d_func(func, x, h=1e-5):
    return (func(x+h)-func(x))/h
    # return (func(x + h) - func(x - h)) / (2 * h)

#------
# Q2-2
#------
def square(x):
    return x**2

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10,10)
y = sigmoid(x)
y_prime = d_func(sigmoid,  x)
plt.plot(x,y)
plt.grid()
plt.show()
plt.plot(x,y_prime)
plt.grid()
plt.show()
```

```python
#----------
# ReLu関数
#----------
class Relu:
    def __init__(self):
        self.mask = None
    def func_out(self, x):   # 関数の定義
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0
        return out
    def func_diff(self, x):  # 関数の微分形
        result = np.ones((x.shape[0], x.shape[1]))
        result[self.mask] = 0
        return result
```

```python
#------
# Q4-1
#------
# A、0以上の時はそのまま、0以下の時は0を返す関数
x = np.linspace(-10,10)
relu = Relu()
y = relu.func_out(x)
plt.plot(x,y)
plt.grid()
plt.show()

#------
# Q4-3
#------
```
